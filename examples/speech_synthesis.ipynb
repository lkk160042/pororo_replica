```python
import IPython
from pororo import Pororo
```

```python
tts = Pororo("tts", lang="multi")
```

```python
# Typical TTS
wave = tts("포로로 많이 사용해주세요!", lang="ko")
IPython.display.Audio(wave, rate=22050)
```

```python
# Cross-lingual Voice Style Transfer
wave = tts("안녕하세요 저는 미국 사람이에요.", lang="ko", speaker="en")
IPython.display.Audio(wave, rate=22050)
```

```python
# Code-Switching
wave = tts("안녕하세요 저는 한국 사람이에요.", lang="ko", speaker="en-18,ko")
IPython.display.Audio(wave, rate=22050)
```
